## Modelling

##Encoding

````python
le = LabelEncoder()

data.accident_severity = le.fit_transform(data.accident_severity)
data.user_sex = le.fit_transform(data.user_sex)
data.lighting = le.fit_transform(data.lighting)
data.intersection = le.fit_transform(data.intersection)
data.atmosphere = le.fit_transform(data.atmosphere)
data.collision = le.fit_transform(data.collision)
data.road_category = le.fit_transform(data.road_category)
data.traff_regime = le.fit_transform(data.traff_regime)
data.pedestrian_action = le.fit_transform(data.pedestrian_action)
data.localisation = le.fit_transform(data.localisation)
data.user_category = le.fit_transform(data.user_category)
data.longitud_profile = le.fit_transform(data.longitud_profile)
data.drawing_plan = le.fit_transform(data.drawing_plan)
data.surface_cond = le.fit_transform(data.surface_cond)
data.acc_situation = le.fit_transform(data.acc_situation)

````

````python
data.head(20)
````

````python
df = data[['intersection',	'atmosphere',	'collision',	'localisation',	'user_category',	'accident_severity',	'user_sex',	'pedestrian_action',	'road_category',	'traff_regime',	'longitud_profile',	'drawing_plan',	'surface_cond',	'acc_situation']]
df = df.dropna()
````

````python
for  col  in  df . columns : 
    df [ col ]  =  df [ col ] . apply ( lambda  x :  int ( x )  )

df . head ()
````

````python
x  =  df. drop (['accident_severity'],  axis = 1 ) 
y  =  df ['accident_severity']
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
````

````python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
````

## Random Forest Classifier

````python
RFC = RandomForestClassifier(random_state = 0)
RFC.fit(x_train, y_train)
predict_RFC = RFC.predict(x_test)
accuracy_score(y_test, predict_RFC)
````

````python
report = classification_report(y_test, predict_RFC)
print(report)
````

## Logistic Regression

````python
LR = LogisticRegression(random_state = 0)
LR.fit(x_train, y_train)
predict_LR = LR.predict(x_test)
accuracy_score(y_test, predict_LR)
````

````python
report = classification_report(y_test, predict_LR)
print(report)
````

## Decision Tree Classifier

````python
DT = DecisionTreeClassifier(min_samples_leaf = 10, max_leaf_nodes = 20)
DT.fit(x_train, y_train)
predict_DT = DT.predict(x_test)
accuracy_score(y_test, predict_DT)
````

````python
report = classification_report(y_test, predict_DT)
print(report)
````

## XGBoost

````python
XGB = xgb.XGBRegressor()
XGB.fit(x_train, y_train)
predict_XGB = XGB.predict(x_test)
xgb_rsme = np.sqrt(Rsquared(y_test, predict_XGB))
print("R Squared for XGBoost: ", np.mean(xgb_rsme))
````

##LightGBM

````python
LGBM = lightgbm.LGBMRegressor()
LGBM.fit(x_train, y_train)
predict_LGBM = LGBM.predict(x_test)
lgbm_rsme = np.sqrt(Rsquared(y_test, predict_LGBM))
print("R Squared for LGBM: ", np.mean(lgbm_rsme))
````
