## Data Cleaning

```python
data = data.rename(columns={'Num_Acc': 'Accident ID',	'catv': 'Category of vehicle', 'obsm': 'collided with',	'num_veh': 'Vehicle Identification'})

data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
# Filling the Accident ID column
data['Accident ID'] = data['Accident ID'].fillna(data['Accident ID'].mode()[0])

# Filling the senc column
data['senc'] = data['senc'].fillna(data['senc'].mode()[0])

# Filling the Category of vehicle column
data['Category of vehicle'] = data['Category of vehicle'].fillna(data['Category of vehicle'].mode()[0])

# Filling the obs column
data['obs'] = data['obs'].fillna(data['obs'].mode()[0])

# Filling the collided with column
data['collided with'] = data['collided with'].fillna(data['collided with'].mode()[0])

# Filling the choc with column
data['choc'] = data['choc'].fillna(data['choc'].mode()[0])

# Filling the manv column
data['manv'] = data['manv'].fillna(data['manv'].mode()[0])

# Filling the Vehicle Identification column
data['Vehicle Identification'] = data['Vehicle Identification'].fillna(data['Vehicle Identification'].mode()[0])
```

```python
vehicles=data[['Accident ID','senc','Category of vehicle', 'obs', 'collided with', 'choc', 'manv', 'Vehicle Identification']]

## Missing Imputation
def missing_imputation(x):
    x=x.fillna(' ')
    return x

vehicles=vehicles.apply(missing_imputation)
```

```python
vehicles.isnull().sum()
```

## Removing duplicate

```python
duplicate = data.duplicated().sum()
if duplicate:
    print('Duplicate rows in Dataset are {}'.format(duplicate))
else:
    print('Dataset contains no duplicate values')
```

```python
for i in data.columns:
  duplicated_data = data[i].duplicated().sum()
  unique_values = data[i].nunique()
  if duplicated_data:
    print('Duplicate rows in dataset at {} column are {} with {} unique items'.format(i, duplicated_data, unique_values))
  else: 
    print('Dataset contains no duplicate values in {}'.format(i))
```

```python
#print(data.shape)
data = data.drop_duplicates()
print(data.shape)
```
